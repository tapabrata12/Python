### Q1 What is rule based system? write a program in python to discuss rule-basedmsystem?
- **Ans**:
A rule-based system is a type of artificial intelligence that uses predefined rules
to make decisions or solve problems. It operates by following a set of "if-then"
statements: when certain conditions (the "if" part) are met, specific actions or
conclusions (the "then" part) are triggered. These systems are typically used in
areas like expert systems, where they simulate human decision-making by
applying domain-specific knowledge encoded in the form of rules.
For example:
• If temperature is above 100°C, then the water is boiling.

```python
import networkx as nx
import matplotlib.pyplot as plt
# Create a directed graph for the rule-based system
G = nx.DiGraph()
# Adding nodes and edges for each rule
rules = [
 ("Symptom", "Fever"),
 ("Symptom", "Cough"),
 ("Symptom", "Headache"),
 ("Fever", "Take rest and fluids"),
 ("Cough", "Take cough syrup"),
 ("Headache", "Take painkillers")
]
G.add_edges_from(rules)
# Draw the graph
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color="lightgreen", node_size=3000,
font_size=10, font_weight="bold")
plt.title("Rule-Based System Visualization")
plt.show()
```

### Q2 What is Semantic Networks in AI? write a program in python to discuss Semantic Networks

- Ans:
Semantic networks in AI are a type of knowledge representation that uses a graph
structure to represent relationships between concepts or entities. In this structure,
nodes represent concepts (like objects, ideas, or events), and edges (or links)
represent the relationships between these concepts (such as "is a," "part of," or
"related to").
Semantic networks help model knowledge in a way that mimics how humans
associate ideas. They are often used in natural language processing, reasoning, and
knowledge-based systems to represent hierarchical or associative knowledge.
For example, a semantic network might represent:

- Dog "is a" Mammal.

- Mammal "is a" Animal.

- Dog "has" Tail.

- This way, AI systems can infer new knowledge by traversing the network and
understanding relationships between different nodes.


```python
import networkx as nx
import matplotlib.pyplot as plt
# Create a semantic network
G = nx.DiGraph()
G.add_edges_from([
('Animal', 'Mammal'),
('Mammal', 'Cat'),
('Mammal', 'Dog'),
('Animal', 'Bird'),
('Bird', 'Sparrow'),
('Bird', 'Eagle'),
('Mammal', 'Human')
])
# Draw the network
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_color='lightblue', node_size=3000,
font_size=10, font_weight="bold")
plt.title("Semantic Network Visualization")
plt.show()
```

### Q3 What is what is Conceptual Graphs? write a program in python to discuss Conceptual Graphs

Ans: -
In AI, conceptual graphs are a way to represent and reason about knowledge
using graph structures. They were developed by John F. Sowa as part of his
work on conceptual structures, which are intended to capture the meaning of
information in a formal and visual manner.

Here’s a breakdown of how conceptual graphs work in AI:
- 1.Nodes (Concepts): Represent entities, objects, or ideas. For example, in a graph
representing a knowledge domain, nodes might represent "Person," "Dog," or
"City."
- 2.Edges (Relations): Define the relationships between concepts. For example, an
edge might represent the relationship "owns" between a "Person" and a "Dog."
- 3.Graphs (Conceptual Structures): The combination of nodes and edges forms a
graph that represents a piece of knowledge. For instance, a graph might show
that "John (Person) owns Rover (Dog)."
- 4.Semantic Interpretation: Conceptual graphs provide a clear, formal way to
encode and interpret the meaning of concepts and their relationships. This
helps in tasks like understanding natural language, reasoning about
information, and making inferences.
- 5.Inference and Reasoning: AI systems can use conceptual graphs to perform
logical reasoning and derive new information from existing knowledge. For
example, if you have a graph indicating that "John owns Rover" and "Rover is a
Dog," you can infer that "John owns a Dog."
Overall, conceptual graphs offer a structured way to represent knowledge that
facilitates reasoning and understanding in AI systems, making them valuable in
fields like knowledge management, natural language processing, and expert
systems.


```python
import networkx as nx
import matplotlib.pyplot as plt
# Create a new graph
G = nx.Graph()
# Add nodes
G.add_node("John", label="Person")
G.add_node("Rover", label="Dog")
# Add edges with relationship labels
G.add_edge("John", "Rover", relationship="owns")
# Define node positions (optional)
pos = nx.spring_layout(G)
# Draw the graph
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue',
font_size=12, font_weight='bold')
edge_labels = nx.get_edge_attributes(G, 'relationship')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
# Show the plot
plt.show()
```
